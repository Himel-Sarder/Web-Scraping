```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time

# Setup
service = Service('E:/Selenium/chromedriver-win64/chromedriver-win64\chromedriver.exe')
driver = webdriver.Chrome(service=service)

# Base URL up to '&page='
base_url = "https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DApple&page="

all_html = ""
total_pages = 24 

for page in range(1, total_pages + 1):
    url = base_url + str(page)
    print(f"Scraping page {page}")
    
    driver.get(url)
    time.sleep(3)  # Wait for the page to load
    
    # Scroll to load full content
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    all_html += driver.page_source

# Save HTML
with open('flipkartMobile.html', 'w', encoding='utf-8') as f:
    f.write(all_html)

driver.quit()
print("✅ Scraping complete.")
```

```python
with open('ChaldalFishs.html', 'r', encoding='utf-8') as f:
    html = f.read() 

soup2 = BeautifulSoup(html, 'lxml')

containers = soup2.find_all('div', {'class': 'product'})


name = []
quantity = []
price = []

for i in containers:
    name.append(i.find('div', {'class':'name'}).text)
    quantity.append(i.find('div', {'class': 'subText'}).text)
    try:
        price.append(i.find('div', {'class': 'discountedPrice'}).text.strip())
    except:
        price.append(i.find('div', {'class': 'price'}).text.strip())
```

```python
import pandas as pd
df = pd.DataFrame({
    'Name':name,
    'quantity':quantity,
    'price':price
})
```

```python
import re

# Remove "After Cutting"
df['Name'] = df['Name'].str.replace(r'After Cutting', '', regex=True)

# Remove patterns like "±50 gm", "±30 gm", etc.
df['Name'] = df['Name'].str.replace(r'±\d+\s*[Gg]m', '', regex=True)

# Remove patterns like "45-55 pcs", "16-20 pcs", etc.
df['Name'] = df['Name'].str.replace(r'\d+\s*-\s*\d+\s*pcs', '', regex=True)

# Clean extra spaces
df['Name'] = df['Name'].str.replace(r'\s+', ' ', regex=True).str.strip()

# Remove "(Special Offer)"
df['Name'] = df['Name'].str.replace(r'\(Special Offer\)', '', regex=True)

# Remove "±50 gm", "± 30 gm", etc. (with or without space after ±)
df['Name'] = df['Name'].str.replace(r'±\s*\d+\s*[Gg]m', '', regex=True)

# Clean extra spaces
df['Name'] = df['Name'].str.replace(r'\s+', ' ', regex=True).str.strip()

# Remove all parentheses and their contents
df['Name'] = df['Name'].str.replace(r'\([^)]*\)', '', regex=True)

# Clean extra spaces
df['Name'] = df['Name'].str.replace(r'\s+', ' ', regex=True).str.strip()
```
